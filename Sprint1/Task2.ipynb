Ниже в блоке по примеру создайте базовый класс Animal и производные классы (2-3 например Dog, Cat и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также статические и обычными атрибуты и метода класса с реализацией взаимодействия между объектами.

```python
public interface IAction
{
    void Action();
}
public abstract class Animal : IAction
{
    private string _name;
    private int _age;
    public static int AnimalCount { get; private set; } = 0;

    public string Name{
        get { return _name;}
        set {
            if (!string.IsNullOrWhiteSpace(value))
            _name = value;
        else
            throw new ArgumentException("Имя не должно быть пустым.");
        }
    }
    public int Age{
        get { return _age;}
        set
        {
            if (value > 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть меньше или равен нулю.");
        }
    }
    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        AnimalCount++;
    }
    public void Hello()
    {
        Console.WriteLine($"Привет! Я - животное. Меня зовут {Name} и мне {Age}");
    }
    public abstract void Action();
    public static void DisplayAnimalCount()
    {
        Console.WriteLine($"Общее количество животных: {AnimalCount}");
    }

}
public class Copybara : Animal
{
    private int _weight;
    public int Weight
    {
        get {return _weight;}
        set {
                if(value > 0)
                    _weight = value;
                else
                throw new ArgumentOutOfRangeException("Вес не может быть меньше или равен нулю.");}
    } 
    public Copybara(string name, int age, int weight) : base(name, age)
    {
        Weight = weight;
    }
    public override void Action()
    {
        Console.WriteLine($"{Name} - Копибара весом {Weight}кг плавает озере.\n");
    }
}
public class Bird : Animal
{
    private int _wingspan;
    public int Wingspan
    {
        get { return _wingspan; }
        set
        {
            if (value > 0)
                _wingspan = value;
            else
                throw new ArgumentOutOfRangeException("Размах крыльев не может быть меньше или равен нулю.");
        }
    }

    public Bird(string name, int age, int wingspan) : base(name, age)
    {
        Wingspan = wingspan;
    }

    public override void Action()
    {
        Console.WriteLine($"{Name} - птица с размахом крыльев {Wingspan} см летает над озером.\n");
    }
}
public class Turtle : Animal
{
    public double _shellLength;
    public double ShellLength
    {
        get { return _shellLength; }
        set {
            if (value > 0 && value <= 0.8)
                _shellLength = value;
            else
                throw new ArgumentException("Длина панциря должна быть больше 0 и не превышать 0.8 метра.");
        }
    }
    public Turtle(string name, int age, double shellLength) : base(name, age)
    {
        ShellLength = shellLength;
    }
    public override void Action()
    {
        Console.WriteLine($"Чeрепаха {Name} ползёт с длинной панцыря {ShellLength}м. \n");
    }
}

Copybara copybara = new Copybara("Арлен", 4, 25);
Animal animal = (Animal)copybara;
animal.Hello();
copybara.Action();

Bird bird = new Bird("Кеша", 1, 50);
Animal animal2 = (Animal)bird;
animal2.Hello();
bird.Action();

Turtle turtle = new Turtle("Тартила", 9, 0.5);
Animal animal3 = (Animal)turtle;
animal3.Hello();
turtle.Action();

Animal.DisplayAnimalCount();
```
