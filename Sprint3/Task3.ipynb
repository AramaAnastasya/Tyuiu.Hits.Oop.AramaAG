Создать базовый класс Person в C#, который будет представлять информацию о человеке. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.
<br><b>Tребования к базовому классу Person:</b>
<br>• Атрибуты: Имя (Name), Возраст (Age), Пол (Gender).
<br>• Методы:
o GetInfo(): метод для получения информации о человеке в виде строки.
o SayHello(): метод для вывода приветствия от человека.
<br><b>Требования к производным классам:</b>
1. Студент (Student): Должен содержать дополнительные атрибуты, такие как
Университет (University) и Курс (Course). Метод SayHello() должен быть
переопределен для добавления информации о курсе при приветствии.
2. Работник (Employee): Должен содержать дополнительные атрибуты, такие
как Компания (Company) и Зарплата (Salary). Метод GetInfo() должен быть
переопределен для включения информации о компании и зарплате.
3. Преподаватель (Teacher) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Предмет (Subject) и Стаж работы
(Experience). Метод SayHello() должен быть переопределен для добавления
информации о преподаваемом предмете при приветствии.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;
using System.Collections.Generic;

public interface IResearcher
{
    void ConductResearch();
}

public class Person
{
    private string _name;
    private int _age;
    private string _gender;
    private string _address;

    public string Name
    {
        get { return _name; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                throw new ArgumentException("Имя не должно быть пустым.");
        }
    }

    public int Age
    {
        get { return _age; }
        set
        {
            if (value > 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть меньше или равен нулю.");
        }
    }

    public string Gender
    {
        get { return _gender; }
        set
        {
            if (value == "Мужчина" || value == "Женщина")
                _gender = value;
            else
                throw new ArgumentException("Пол человека должен быть: мужчина или женщина.");
        }
    }

    public string Address
    {
        get { return _address; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _address = value;
            else
                throw new ArgumentException("Адрес не должен быть пустым.");
        }
    }

    public Person(string name, int age, string gender, string address)
    {
        Name = name;
        Age = age;
        Gender = gender;
        Address = address;
    }

    public virtual void SayHello()
    {
        Console.WriteLine($"Привет! Я - {Gender}. Меня зовут {Name} и мне {Age} лет.");
    }

    public void GetInfo()
    {
        Console.WriteLine($"Получение информации:\nИмя: {Name}\nВозраст: {Age}\nПол: {Gender}\nАдрес: {Address}");
    }

    public void ChangeAddress(string newAddress){
        Address = newAddress;
        Console.WriteLine($"Адрес изменен на: {Address}");
    }
}

public class Student : Person
{
    private string _university;
    private int _course;
    private double _gpa;

    public string University
    {
        get { return _university; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _university = value;
            else
                throw new ArgumentException("Название университета не должно быть пустым.");
        }
    }

    public int Course
    {
        get { return _course; }
        set
        {
            if (value >= 1 && value <= 6)
                _course = value;
            else
                throw new ArgumentException("Курс не должен быть меньше 1 и не больше 6.");
        }
    }

    public double GPA
    {
        get { return _gpa; }
        set
        {
            if (value > 0)
                _gpa = value;
            else
                throw new ArgumentException("Средний балл не может быть меньше 0.");
        }
    }

    public Student(string name, int age, string gender, string address, string university, int course, double gpa)
        : base(name, age, gender, address)
    {
        University = university;
        Course = course;
        GPA = gpa;
    }

    public override void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name} и мне {Age} лет. Я {Gender}, учусь в университете {University} на {Course} курсе.");
    }

    public void GetInfo()
    {
        Console.WriteLine($"Получение информации:\nИмя: {Name}\nВозраст: {Age}\nПол: {Gender}\nАдрес: {Address}");
    }

    public void GetInfo(bool detailed)
    {
        if (detailed)
        {
            Console.WriteLine($"Получение детальной информации:\nИмя: {Name}\nВозраст: {Age}\nПол: {Gender}\nАдрес: {Address}\nУниверситет: {University}\nКурс: {Course}\nСредний балл: {GPA}");
        }
        else
        {
            base.GetInfo();
        }
    }

    public void CalculateGPA()
    {
        Console.WriteLine($"Я, {Name}, рассчитала мой средний балл: {GPA}");
    }
}

public class Teacher : Person, IResearcher
{
    private string _subject;
    private double _experience;
    private int _numberOfStudents;

    public string Subject
    {
        get { return _subject; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _subject = value;
            else
                throw new ArgumentException("Название учебной дисциплины не должно быть пустым.");
        }
    }

    public double Experience
    {
        get { return _experience; }
        set
        {
            if (value > 0)
                _experience = value;
            else
                throw new ArgumentException("Стаж преподавания не может быть меньше 0.");
        }
    }

    public int NumberOfStudents
    {
        get { return _numberOfStudents; }
        set
        {
            if (value >= 0)
                _numberOfStudents = value;
            else
                throw new ArgumentException("Количество студентов не может быть отрицательным.");
        }
    }

    public Teacher(string name, int age, string gender, string address, string subject, int experience, int numberOfStudents)
        : base(name, age, gender, address)
    {
        Subject = subject;
        Experience = experience;
        NumberOfStudents = numberOfStudents;
    }

    public override void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}, мне {Age} лет. Я веду предмет {Subject} со стажем работы: {Experience} лет.");
    }

    public void AddStudent()
    {
        NumberOfStudents++;
        Console.WriteLine($"Добавлен новый студент. Общее количество студентов: {NumberOfStudents}");
    }

    public void ConductResearch()
    {
        Console.WriteLine($"Я, {Name}, провожу исследования в области {Subject}.");
    }
}

public class GraduateStudent : Student
{
    private string _thesisTopic;

    public string ThesisTopic
    {
        get { return _thesisTopic; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _thesisTopic = value;
            else
                throw new ArgumentException("Тема диссертации не должна быть пустой.");
        }
    }

    public GraduateStudent(string name, int age, string gender, string address, string university, int course, double gpa, string thesisTopic)
        : base(name, age, gender, address, university, course, gpa)
    {
        ThesisTopic = thesisTopic;
    }

    public void DefendThesis()
    {
        Console.WriteLine($"Я, {Name}, защищаю диссертацию на тему: {ThesisTopic}.");
    }
}

public class PersonCollection<T> where T : Person
{
    private List<T> _people = new List<T>();

    public void Add(T person)
    {
        _people.Add(person);
    }

    public void Remove(T person)
    {
        _people.Remove(person);
    }

    public void DisplayPeople()
    {
        foreach (var person in _people)
        {
            person.SayHello();
        }
    }
}

PersonCollection<Person> people = new PersonCollection<Person>();
people.Add(new Person("Василина", 35, "Женщина", "ул. Ленина, 123"));
people.Add(new Student("Ярослава", 22, "Женщина", "ул. Гагарина, 78", "ТИУ", 2, 76.7));
people.Add(new Teacher("Алексей", 40, "Мужчина", "ул. Советская, 10", "Базы данных", 10, 30));
people.Add(new GraduateStudent("Мария", 25, "Женщина", "ул. Школьная, 5", "МГУ", 6, 85.0, "Искусственный интеллект"));

people.DisplayPeople();
Console.WriteLine("\n");
```

```
using System;
using System.Collections.Generic;

public interface IInstallable
{
    void Install();
}

public abstract class Component : IInstallable
{
    public string Name { get; protected set; }
    public double Price { get; protected set; }

    public Component(string name, double price)
    {
        Name = name;
        Price = price;
    }


    public virtual string GetInfo()
    {
        return $"Компонент: {Name}, Стоимость: {Price}";
    }

    public override string ToString()
    {
        return GetInfo();
    }

    public virtual void Install()
    {
        Console.WriteLine($"Установка {Name}...");
    }

    public static bool operator ==(Component c1, Component c2)
    {
        if (ReferenceEquals(c1, c2)) return true;
        if (ReferenceEquals(c1, null) || ReferenceEquals(c2, null)) return false;
        return c1.Name == c2.Name && c1.Price == c2.Price;
    }

    public static bool operator !=(Component c1, Component c2)
    {
        return !(c1 == c2);
    }

    public override bool Equals(object obj)
    {
        if (obj is Component other)
        {
            return this == other;
        }
        return false;
    }

    public override int GetHashCode()
    {
        return (Name, Price).GetHashCode();
    }
}

public class CPU : Component
{
    public CPU(string name, double price) : base(name, price) { }

    public override string GetInfo()
    {
        return $"CPU: {Name}, Price: {Price}";
    }
}

public class RAM : Component
{
    public RAM(string name, double price) : base(name, price) { }

    public override string GetInfo()
    {
        return $"RAM: {Name}, Стоимость: {Price}";
    }
}

public class GPU : Component
{
    public GPU(string name, double price) : base(name, price) { }

    public override string GetInfo()
    {
        return $"GPU: {Name}, Стоимость: {Price}";
    }
}

public class HDD : Component
{
    public HDD(string name, double price) : base(name, price) { }

    public override string GetInfo()
    {
        return $"HDD: {Name}, Стоимость: {Price}";
    }
}

public class Computer
{
    private List<Component> components = new List<Component>();

    public void AddComponent<T>(T component) where T : Component, IInstallable
    {
        components.Add(component);
        component.Install();
    }

    public void RemoveComponent<T>(T component) where T : Component
    {
        components.Remove(component);
    }

    public double GetTotalPrice()
    {
        double totalPrice = 0;
        foreach (var component in components)
        {
            totalPrice += component.Price;
        }
        return totalPrice;
    }

    public void DisplayComponents()
    {
        foreach (var component in components)
        {
            Console.WriteLine(component.GetInfo());
        }
    }
}

public static class ComputerFactory
{
    public static Computer CreateBasicComputer()
    {
        var computer = new Computer();
        computer.AddComponent(new CPU("Intel Core i5", 200));
        computer.AddComponent(new RAM("8GB DDR4", 50));
        computer.AddComponent(new HDD("1TB HDD", 60));
        return computer;
    }

    public static Computer CreateGamingComputer()
    {
        var computer = new Computer();
        computer.AddComponent(new CPU("Intel Core i7", 300));
        computer.AddComponent(new RAM("16GB DDR4", 80));
        computer.AddComponent(new GPU("NVIDIA RTX 3080", 700));
        computer.AddComponent(new HDD("1TB SSD", 100));
        return computer;
    }
}

// Создание базового компьютера
var basicComputer = ComputerFactory.CreateBasicComputer();
Console.WriteLine("Базовый компонент компьютера:");
basicComputer.DisplayComponents();
Console.WriteLine($"Total Price: {basicComputer.GetTotalPrice()}");

// Создание игрового компьютера
var gamingComputer = ComputerFactory.CreateGamingComputer();
Console.WriteLine("\nИгровой компонент компьютера:");
gamingComputer.DisplayComponents();
Console.WriteLine($"Total Price: {gamingComputer.GetTotalPrice()}");

// Сравнение компонентов
var cpu1 = new CPU("Intel Core i5", 200);
var cpu2 = new CPU("Intel Core i5", 200);
Console.WriteLine($"CPU1 == CPU2: {cpu1 == cpu2}");
Console.WriteLine($"CPU1 != CPU2: {cpu1 != cpu2}");
```
